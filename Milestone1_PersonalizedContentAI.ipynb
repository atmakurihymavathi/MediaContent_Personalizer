```python
import pandas as pd
import random
import json
```

```python
data = []
with open("News_Category_Dataset_v3.json", 'r', encoding='utf-8') as f:
    for line in f:
        try:
            data.append(json.loads(line.strip()))
        except json.JSONDecodeError:
            continue  # Skip incorrect lines

df = pd.DataFrame(data)
df.head()
```

```python
def get_recommendations(user_interest, n=5):
    filtered = df[df['category'].str.contains(user_interest, case=False, na=False)]

    if len(filtered) == 0:
        return "No matching content found. Try a different interest!"

    return filtered.sample(min(n, len(filtered)))[['headline', 'short_description']]
```

```python
user_input = input("Enter your interest (e.g., Tech, Sports, Politics): ")
get_recommendations(user_input)
```
